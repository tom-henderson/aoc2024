```python
input_file = "day1.input"
list1 = []
list2 = []
with open(file=input_file, mode="r") as input:
    for line in input.readlines():
        a, b = line.split("   ")
        list1.append(int(a))
        list2.append(int(b))

assert list1[0] == 77221
assert list1[-1] == 19976
assert list2[0] == 93653
assert list2[-1] == 46609
assert len(list1) == 1000
assert len(list2) == 1000
```

```python
list1.sort()
list2.sort()
print(list1[0])
print(list2[0])
```

```python
diffs = []
for i in range(len(list1)):
    diffs.append(max(list1[i], list2[i]) - min(list1[i], list2[i]))
```

```python
total = 0
for i in diffs:
    total += i

print(total)
assert total == 1580061 # ⭐️
```

```python
num_times = {}
for i in list2:
    if num_times.get(i):
        num_times[i] = num_times[i] + 1
    else:
        num_times[i] = 1

num_times
```

```python
sim_score = 0
for i in list1:
    sim_score = sim_score + (i * num_times.get(i, 0))

print(sim_score)
assert sim_score == 23046913 # ⭐️
```
